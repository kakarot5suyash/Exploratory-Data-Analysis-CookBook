# Dropping Data rows and columns in a dataset

## Importing pandas

```python
import pandas as pd
```

## Loading dataset and including relvant columns

```python
marketing_data = pd.read_csv('Data/marketing_campaign.csv')
marketing_data = marketing_data[['ID', 'Year_Birth','Kidhome', 'Teenhome']]
```

```python
marketing_data.head(10)
```

## Delete a specified row at index value 1

```python
marketing_data.drop(labels = [1], axis = 0)
```

## Delete a single column

```python
marketing_data.drop(labels =['Year_Birth'], axis = 1)
```
